```python
m = geemap.Map()

# New region covering the entire world
worldRegion = ee.Geometry.Polygon(
  [[[-180, -90], [180, -90], [180, 90], [-180, 90], [-180, -90]]],
  None, False
)

Vis = {
  'min': -670,
  'max': 576,
  'palette': ['1a3678', '2955bc', '5699ff', '8dbae9', 'acebff', 'e5f9ff', 'fdffb4', 'ffe692', 'ffc969', 'ffc969', 'ffa12d', 'ca531a', 'ff0000', 'ab0000']
}

# Function to download and display data for a specific year
def downloadAndDisplayData(year, description):
    datasetpre = ee.ImageCollection("IDAHO_EPSCOR/TERRACLIMATE") \
    .select('tmmx') \
    .filter(ee.Filter.date(year + '-12-01', year + '-12-31'))

    def func_aao(image):
          return image.multiply(0.1)
    datasetpre = datasetpre.map(func_aao).median().clip(geometry2)

    m.addLayer(datasetpre, Vis, description)
    print(type(datasetpre))
    img = ee.Image(datasetpre)
    
    geemap.ee_export_image(
    img, "Terraclim.tif", scale=10000, region=geometry2, file_per_band=False
)
  
m
```

```python
# Loop through each year from 1985 to 2022
for year in range(1985, 2023, 1):
  description = str(year)
  downloadAndDisplayData(str(year), description)

m
```
